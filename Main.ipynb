FEATURE SELECTION PART

```python
movies.columns
```

```python
movies = movies[['id','title','genres','overview','release_date','cast', 'director', 'poster_path','original_language']]
```

```python
movies
```

```python
movies['tags'] = movies['genres'] + movies['overview'] + movies['cast'] + movies['director'] + movies['original_language']
```

```python
movies
```

```python
new_data = movies.drop(columns=['genres','overview','cast','director','original_language'])
```

```python
new_data
```

Feature Extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=5000, stop_words='english', ngram_range=(1,2))
```

```python
vectors = tfidf.fit_transform(new_data['tags'].values.astype('U'))
```

```python
vectors.shape
```

```python
pickle.dump(
    new_data,
    open('movies_list.pkl', 'wb')
)

joblib.dump(
    vectors,
    'vectors.pkl',
    compress=3
)

joblib.dump(
    tfidf,
    'vectorizer.pkl',
    compress=3
)
```

```python
pickle.load(open('movies_list.pkl','rb'))
```

Recommend Function

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
def recommend(movie):

    # FIND MOVIE INDEX
    movie_index = new_data[new_data['title'] == movie].index[0]

    # GET VECTOR OF SELECTED MOVIE
    movie_vector = vectors[movie_index]

    # COMPUTE SIMILARITY DYNAMICALLY
    similarity = cosine_similarity(movie_vector, vectors).flatten()

    # GET TOP SIMILAR MOVIES
    distances = sorted(
        list(enumerate(similarity)),
        reverse=True,
        key=lambda x: x[1]
    )[1:11]

    # STORE RESULTS
    recommended_movies = []

    for i in distances:
        recommended_movies.append(new_data.iloc[i[0]].title)

    return recommended_movies
```

```python
recommend("Interstellar")
```

```python
recommend("3 Idiots")
```
